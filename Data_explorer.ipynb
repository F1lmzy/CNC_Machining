```python
def print_matrix_info(vibration_data):
    """Helper function to print formatted matrix information

    Args:
        vibration_data: Matrix of vibration measurements
    """
    print("Matrix Shape:", vibration_data.shape)
    print("\nFirst 5 rows of data:")
    print("X-axis | Y-axis | Z-axis")
    print("-" * 40)
    for i in range(len(vibration_data)):
        print(" | ".join(f"{val:.4f}" for val in vibration_data[i]))

file = Path("./data/M01/OP00/good/M01_Aug_2019_OP00_000.h5").absolute()
vib_matrix = data_loader_utils.datafile_read(file, plotting=False)
print_matrix_info(vib_matrix)
```
